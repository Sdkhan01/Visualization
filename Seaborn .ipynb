```python
import seaborn as sns
```

```python
iris  = sns.load_dataset('iris') #Dataset of the flower
```

```python
iris
```

```python
#making the observation by using graph that :
#if sepal lengh increment will increase sepal width or patal width or anything

sns.scatterplot(x = iris.sepal_length, y = iris.sepal_width)
```

```python
#Checking sepal length with respect to petal length

sns.scatterplot(x = iris.sepal_length , y = iris.petal_length)
```

```python
sns.scatterplot(x = iris.petal_width , y = iris.petal_length)
```

```python
sns.displot(iris)
```

```python
#Distribution plot in seaborn same as histogram in matlot lib

sns.displot(iris['sepal_length'])
```

```python
#Loading new data set tips

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#find if customer with higher bill giving higher tip

sns.scatterplot(x = tips.total_bill, y = tips.tip)
```

```python
tips['smoker'].count()
```

```python
#Findign number of both smoker and non smoker

tips['smoker'].value_counts()
```

```python
#finding smoker details using plotting
sns.relplot( x = tips.total_bill, y = tips.tip, data = tips , hue = 'smoker' )
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, data = tips , hue = 'sex')
```

```python
#Finding details about the customer that they came alone or in groups 

sns.relplot(x = tips.total_bill , y = tips.tip , data= tips, hue = 'size' )
```

```python
#To find out how many customers came alone,or in group with exact number

tips['size'].value_counts()
```

```python
#Finding details about the customer that they came alone or in groups and indicate them with different symbols
sns.relplot(x = tips.total_bill , y = tips.tip ,data = tips, style = 'size' , hue = 'sex')
```

```python
#Checking how many male and female customer come in dnner and lunch with respective data

sns.relplot(x = tips.total_bill, y = tips.tip, data = tips, style = 'sex', hue = 'time')
```

```python
#which day we are getting the higher bill 
#Catagorical data

sns.catplot(x = 'day' , y = 'total_bill', data = tips )
```

```python
#Who is paying more tip, smoker or non smoker

sns.catplot(x = 'smoker', y = 'tip' , data = tips)
```

```python
#Using joinplot function 

sns.jointplot(x = tips.total_bill, y = tips.tip)
```

```python
#It will show every one vs everyone
#Will plot every numerical with each other

sns.pairplot(tips)
```
