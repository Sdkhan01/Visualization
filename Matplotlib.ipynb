```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#Generating the data 
#From 0 to 10 in 200 parts
x = np.linspace(0,10,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
#sin wave
plt.plot(x,y)
```

```python
#Labeling or decorating the x axis data and y axis data
plt.plot(x,y)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my report title")
plt.show()
```

```python
#Generating new data

x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y) #Will plot the dots whereas the x and y axis are meeting
```

```python
#Setting random colors for each plotted dots

color = np.random.rand(50)
sizes = np.random.rand(50)
plt.scatter(x,y , c = color , s = sizes*600,alpha = .6 ) 
# to change size,multiplyin by big no. else dots aren't visible
#alpha is use to manage the opacity of the dots
#c is used to change the colors

#puting x and y lebel
plt.xlabel("This is the x axis data")
plt.ylabel('This is the y label data')
plt.title('This is my first plot')
```

```python
#Can also represent the 3d data

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure() #creating figure
ax = fig.add_subplot(projection = '3d') #adding everything in 3d in the created figure
ax.scatter(x,y,z) #Putted x,y,z along the axis
plt.show()
```

```python
#creating 3d plot

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

fig = plt.figure(facecolor = '#B5CD29')
axis = fig.add_subplot(projection = '3d')
axis.scatter(x,y,z)
```
